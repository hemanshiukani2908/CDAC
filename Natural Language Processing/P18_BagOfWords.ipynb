### By Normal method

```python
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt_tab')
```

```python
sent_1 = "It is a good practice for us."
sent_2 = "It was also good to know about it."
```

```python
t1 = word_tokenize(sent_1)
t2 = word_tokenize(sent_2)
```

```python
# t3 =  t1 + t2
# t3
```

```python
corpus = []
document1 = []
for i in t1:
    corpus.append(i.lower())
    document1.append(i.lower())
```

```python
document2 = []
for i in t2:
    corpus.append(i.lower())
    document2.append(i.lower())
```

```python
voca = set(corpus)
voca
```

```python
voca_lst = list(voca)
voca_lst
```

```python
sent_1_count = [document1.count(word) for word in voca_lst]
sent_2_count = [document2.count(word) for word in voca_lst]
```

```python
# sent_1 = "It is a good practice for us."
# sent_2 = "It was also good to know about it."

df = pd.DataFrame([sent_1_count,sent_2_count],index=[1,2],columns = voca_lst)
```

```python
df
```

## By using in-built library

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvt = CountVectorizer()
```

```python
x_new = cvt.fit_transform([sent_1,sent_2])
```

```python
x_new.toarray()
```

```python
cvt.get_feature_names_out()
```

```python
df = pd.DataFrame(data=x_new.toarray(), columns=cvt.get_feature_names_out())
```

```python
df
```

```python
new = "It was good for us."
```

```python
new_features = cvt.transform([new])
```

```python
new_features.toarray()
```

## Bagofword with N-Grams

```python
ngram = CountVectorizer(ngram_range=(1,2))
```

```python
new_ng = ngram.fit_transform([sent_1,sent_2])
```

```python
ndf = pd.DataFrame(data=new_ng.toarray(), columns=ngram.get_feature_names_out())
```

```python
ndf
```

```python
ngram = CountVectorizer(ngram_range=(2,2))
new_ng = ngram.fit_transform([sent_1,sent_2])
ndf = pd.DataFrame(data=new_ng.toarray(), columns=ngram.get_feature_names_out())
ndf
```

# The TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x_new = tfidf.fit_transform([sent_1,sent_2])
```

```python
x_new.toarray()
```

```python
tfdf = pd.DataFrame(data=x_new.toarray(),index=["Sent1","Sent2"], columns=tfidf.get_feature_names_out())
tfdf
```
