````python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import re
````

## <b>```Open url using urllib.request module```</b>

````python
# read page url

httpObj = urlopen("https://en.wikipedia.org/wiki/Rajgad_Fort")
html_data = httpObj.read()
# html_data     # these give
````

## <b>```Beautify the html data by BeautifulSoup```</b>

````python
soup_data = soup(html_data,'html.parser')
soup_data
````

````python
soup_data.find('title').text
````

## <b>``` Get Relevant text from these page```</b>

````python
data = soup_data.get_text(strip=True)
````

````python
data
````

## <b>```Tokenization```</b>

````python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
````

````python
tokens = word_tokenize(data)
print(tokens)
````

### **Top 10 Frequent word**

````python
from nltk import FreqDist
````

````python
freq = FreqDist(tokens)
freq.plot(10)
````

## <b>```Data Preprocessing```</b>

````python
from nltk.corpus import stopwords
````

````python
tokens1 = [token for token in tokens if token.isalpha() or token.isdigit()]  # remove other value other than alphabetic and digit
print(tokens1)
````

````python
swords = stopwords.words('english')
tokens2 = [word.lower() for word in tokens1 if word.lower() not in swords]
````

````python
freq = FreqDist(tokens2)
freq.plot(20)
````

````python
freq.most_common(5)
````

````python
url = urlopen("https://en.wikipedia.org/wiki/Rajgad_Fort")
data = url.read()
soup = soup(data, 'html.parser')
text = soup.get_text()
tokens = word_tokenize(text)
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
freq = FreqDist(tokens2)
freq.plot(20)
freq.most_common(5)
````

## **Apply Stemming so that we can get more meaning info**

````python
from nltk.stem import PorterStemmer
````

````python
ls = PorterStemmer()
````

````python
url = urlopen("https://en.wikipedia.org/wiki/Tiger")
data = url.read()
soup = BeautifulSoup(data,'html.parser')
text = soup.get_text()
tokens = word_tokenize(text)
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
tokens3 = [ls.stem(token) for token in tokens2 if len(token) > 1]
freq = FreqDist(tokens3)
freq.plot(20)
freq.most_common(5)
````
