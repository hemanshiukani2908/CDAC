## method :1

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import sent_tokenize
```

```python
tfidf = TfidfVectorizer()
```

```python
with open("t19_data.txt",'r') as f:
    lines = f.read()
```

```python
tk = sent_tokenize(lines)
```

```python
tk
```

```python
x_new = tfidf.fit_transform(tk)
```

```python
# x_new.toarray()
```

```python
tfdf = pd.DataFrame(data=x_new.toarray(), columns=tfidf.get_feature_names_out())
tfdf
```
