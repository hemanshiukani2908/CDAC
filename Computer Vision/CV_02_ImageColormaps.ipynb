```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../datasets/images/sachin.jpg', 0)
```

```python
img.shape
```

```python
plt.imshow(img, cmap='Blues')
```

```python
orig_map = plt.colormaps['Blues']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap = reversed_map)
```

```python
orig_map = plt.colormaps['inferno']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap = reversed_map)
```

```python
orig_map = plt.colormaps['Reds']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap = reversed_map)
```

```python
orig_map = plt.colormaps['Greens']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap = reversed_map)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,4,1);
plt.title('Blues')
plt.imshow(img, cmap= plt.colormaps['Blues'].reversed())
plt.subplot(2,4,2);
plt.title('Greens')
plt.imshow(img, cmap= plt.colormaps['Greens'].reversed())
plt.subplot(2,4,3);
plt.title('Reds')
plt.imshow(img, cmap= plt.colormaps['Reds'].reversed())
plt.subplot(2,4,4);
plt.title('Oranges')
plt.imshow(img, cmap= plt.colormaps['Oranges'].reversed())
plt.subplot(2,4,5);
plt.title('Purples')
plt.imshow(img, cmap= plt.colormaps['Purples'].reversed())
plt.subplot(2,4,6);
plt.title('Greys')
plt.imshow(img, cmap= plt.colormaps['Greys'].reversed())
plt.subplot(2,4,7);
plt.title('YlOrBr')
plt.imshow(img, cmap= plt.colormaps['YlOrBr'].reversed())
plt.subplot(2,4,8);
plt.title('OrRd')
plt.imshow(img, cmap= plt.colormaps['OrRd'].reversed())
```
