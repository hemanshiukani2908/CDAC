````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
````

### Load image using CV2 and Plt

````python
img_plt = plt.imread("../datasets/images/jungle.jpg")

plt.imshow(img_plt);
plt.title("Jungle Image");
````

````python
img_cv2 = cv2.imread("../datasets/images/jungle.jpg")
````

````python
cv2.imshow("Jungle Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ```Create Rectangle```
      - Using OpenCv

````python
# help(cv2.rectangle)
````

````python
cv2.rectangle(img_cv2,pt1=(50,200),pt2=(250,400),color=(0,0,255),thickness=2)
cv2.rectangle(img_cv2,pt1=(400,200),pt2=(620,300),color=(0,0,255),thickness=2)
cv2.rectangle(img_cv2,pt1=(230,230),pt2=(380,280),color=(0,0,255),thickness=2)
````

````python
cv2.imshow("Jungle Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ```Create Circle```
      - Using OpenCv

````python
cv2.circle(img_cv2,center=(550,100),radius=50,color=(255,0,0),thickness=2)
````

````python
cv2.imshow("Jungle Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ```Create Line```
     - Using OpenCv

````python
cv2.line(img_cv2,pt1=(240,0),pt2=(240,450),color=(0,255,0),thickness=2)
cv2.line(img_cv2,pt1=(370,0),pt2=(370,450),color=(0,255,0),thickness=2)
````

````python
cv2.imshow("Jungle Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ```Create Text```
     - Using OpenCv

````python
cv2.putText(img_cv2,org=(100,450),fontScale=2,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text="Grass",fontFace=cv2.FONT_ITALIC)
````

````python
cv2.imshow("Jungle Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ```Convert that image to RGB format bcz plt only accept RGB format```

````python
img_RGB_CV2 = cv2.cvtColor(img_cv2,cv2.COLOR_BGR2RGB)
````

````python
plt.imshow(img_RGB_CV2);
````
