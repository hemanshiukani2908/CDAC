## 2. Converting to another color
    - matplotlib accept RGB image format by default
    - cv2(opencv) accept BGR image format by default

````python
img_cv2 = cv2.imread("..\datasets\images\dog.jpg")
````

````python
cv2.imshow("Dog Image",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
img_cv2.shape    # 3 means image in mode BGR format
````

````python
plt.imshow(img_cv2);   # it show BGR format but we require RGB format for matplotlib
````

### BGR ------->RGB

````python
'''
Now, it converted into RGB color format

'''
col_rgb_img = cv2.cvtColor(img_cv2,cv2.COLOR_BGR2RGB)  
````

````python
plt.imshow(col_rgb_img);
````

### BGR -------> GRAYSCALE

````python
img_gray = cv2.imread("../datasets/images/dog.jpg",0)
````

````python
cv2.imshow("Grayscale dog image",img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# plt.imshow(img_gray);

plt.imshow(img_gray,cmap="gray");
````
