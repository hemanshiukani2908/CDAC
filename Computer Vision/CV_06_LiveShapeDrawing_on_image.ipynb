```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load the Image

```python
img_cv2 = cv2.imread("../datasets/images/rajgad.png")

cv2.imshow("Rajgad Hill",img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(img_cv2,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb);
```

```python
def draw(event,x,y,flags,params):
    
    if event == 0:   # mouse moved
#         print("mouse moved")
        pass

    if event == 4:   # mouse up clicked (cursor take out from point clicked)
#         print("mouse up moved")
        pass

    if event == 1:   # mouse down clicked  (cursor clicked)
        cv2.circle(img_cv2,center=(x,y),radius=50,color=(255,0,0),thickness=2)

cv2.namedWindow(winname = "Rajgad HILL")
cv2.setMouseCallback("Rajgad HILL",draw)

while True:
    cv2.imshow("Rajgad HILL",img_cv2)
    
    # window don't be close untill i press x for exit
    if cv2.waitKey(1) & 0xFF == ord('x'):  
        break
        
cv2.destroyAllWindows()
```

```python
flag = False
ix = -1
iy = -1

def draw(event,x,y,flags,params):
    
    global flag,ix,iy
    
    if event == 1:   # mouse down clicked  (cursor clicked)
        flag = True
        ix = x
        iy = y            
            
    elif event == 4:
        cv2.rectangle(img_cv2,pt1=(ix,iy),pt2=(x,y),color=(0,0,255),thickness=2)

    

cv2.namedWindow(winname = "Rajgad HILL")
cv2.setMouseCallback("Rajgad HILL",draw)

while True:
    cv2.imshow("Rajgad HILL",img_cv2)
    
    # window don't be close untill i press x for exit
    if cv2.waitKey(1) & 0xFF == ord('x'):  
        break
        
cv2.destroyAllWindows()
```
