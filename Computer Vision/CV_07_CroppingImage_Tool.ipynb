```python
flag = False
ix = -1
iy = -1

def draw(event,x,y,flags,params):
    
    global flag,ix,iy
    
    if event == 1:   # mouse down clicked  (cursor clicked)
        flag = True
        ix = x
        iy = y            
            
    elif event == 4:
#         fx = x
#         fy = y
        cv2.rectangle(img_cv2,pt1=(ix,iy),pt2=(x,y),color=(0,0,0),thickness=1)

        # crop tools
        cropped = img_cv2[iy:y,ix:x]
        cv2.imwrite("Apple_cropped.jpg",cropped)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

cv2.namedWindow(winname = "Apples Image")
cv2.setMouseCallback("Apples Image",draw)

while True:
    cv2.imshow("Apples Image",img_cv2)
    
    # window don't be close untill i press x for exit
    if cv2.waitKey(1) & 0xFF == ord('x'):  
        break
        
cv2.destroyAllWindows()
```
